```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from IPython.display import SVG, display
from pandas.plotting import parallel_coordinates
from scipy import stats
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
howell=pd.read_csv("/content/howell.csv" , delimiter=";")
howell.head()
```

```python
age_18_exclude = howell["age"] > 18
howell[age_18_exclude].plot(kind = "scatter", x = "weight", y = "height");
```

```python
height = howell["height"]
weight = howell["weight"]
```

```python
weight_18 = weight[age_18_exclude].values.reshape(-1, 1)
height_18 = height[age_18_exclude].values
model = LinearRegression()
model.fit(weight_18, height_18)
α = model.intercept_
β = model.coef_[0]
μ = model.predict(weight_18)
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(weight[age_18_exclude], height[age_18_exclude], "C0.", label='Data')
ax.plot(weight[age_18_exclude], μ, c="k", label='Fit')
sns.regplot(x=weight[age_18_exclude], y=height[age_18_exclude], ax=ax, ci=98, scatter=False, color='gray')
fig.suptitle("Weight vs Height fit and confidence interval")
ax.set_xlabel("Weight")
ax.set_ylabel("Height")
ax.legend()

plt.show()
```

By visually inspecting the fit, the average parameters look quite good, and the 98% interval of the posterior predictive checks covers most of the distribution. Overall, a linear fit appears to be an excellent choice for the relationship between height and weight in people over 18.
